```python
# Data
df = pd.DataFrame(np.random.randn(100, 4), columns= "A B C D".split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [32, 43,50]})
```

```python
df2
```

```python
df.iplot();
```

```python
# iPlot Scatter Plot
df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
# iPlot Bar Plot
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
# Using aggregate functions with iPlot.
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

```python
# iPlot Box Plot
df.iplot(kind='box')
```

## 3D Plots with iPlot

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z':[500,400,300,200,100]})
```

```python
df3.iplot(kind='surface')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z':[5,4,3,2,1]})
df3.iplot(kind='surface')
```

```python
# Changing the color scale
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z':[5,4,3,2,1]})
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
# iPlot Histograms One column
df['A'].iplot(kind='hist', bins=50)
```

```python
# iPlot Histogram ALL columns. You can turn toggle each col individually by clicking on the legend.
df.iplot(kind='hist', bins=50)
```

```python
# iPlot Spread Plot. This is good for stock data!
df[['A', 'B']].iplot(kind='spread')
```

```python
# iPlot bubble plot. Good for United Nations-like Reports
df.iplot(kind='bubble', x='A', y='B', size='C', color='orange')
```
